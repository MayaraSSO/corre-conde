### 1. **Introdução e Contextualização**

Este projeto consiste no desenvolvimento de um jogo digital do gênero **plataforma 2.5D**, desenvolvido como requisito avaliativo para a disciplina de Computação Gráfica. O objetivo central é a aplicação prática de conceitos fundamentais da área, como transformações geométricas, modelos de iluminação, projeções e tratamento de colisões, integrando-os em uma experiência interativa e funcional.

####  1.1. O Projeto: "Corre Conde"

A temática escolhida para este trabalho é a fuga de um vampiro ("Conde") da luz solar. Diferente de jogos de plataforma tradicionais, o foco aqui é a sobrevivência em um ambiente onde a iluminação não é apenas estética, mas uma mecânica punitiva central. O personagem deve navegar por cenários urbanos e catacumbas, utilizando elementos do cenário para projetar sombras e evitar a exposição direta aos feixes de luz solar.

####  1.2. Objetivos da Etapa 1

Nesta fase inicial de desenvolvimento, o foco reside na estabilização do **Protótipo Técnico**. As metas estabelecidas e cumpridas nesta etapa incluem:

* Implementação da base de movimentação horizontal e sistema de salto;

* Desenvolvimento de um motor físico básico para detecção de colisões com o cenário;

* Configuração de uma câmera dinâmica que acompanha o deslocamento do protagonista;

* Criação de um bloco jogável preliminar com obstáculos e plataformas para validação das mecânicas;

* Implementação antecipada de recurso gráfico extra: Sistema de névoa fria noturna com efeito Parallax e interpolação de opacidade (Alpha Blending).

* **Compilador:** **GCC (GNU Compiler Collection)**. Utilizado para a compilação do código-fonte através do terminal Linux com o uso das *flags* `-lGL -lGLU -lglut`.

### 3. **Transformações Geométricas e Interação Espacial**

Nesta etapa, o motor do jogo foi estruturado com foco na manipulação do sistema de coordenadas virtuais e na aplicação de transformações geométricas contínuas, que são os pilares da Computação Gráfica bidimensional.

### 3.1. Animação e Matrizes de Translação (Movimento e Salto)

No contexto do pipeline gráfico, a movimentação do Conde e a aplicação da gravidade não são tratadas apenas como grandezas físicas, mas sim como **Transformações de Translação**.

A cada ciclo de renderização (frame), o sistema calcula os novos deslocamentos nos eixos X e Y. Em vez de utilizar funções complexas de matrizes para cada frame, o motor atualiza diretamente as coordenadas espaciais que compõem os vértices (`glVertex2f`) dos polígonos do personagem (`GL_QUADS`). O pulo é renderizado como uma translação positiva no eixo Y que sofre um decaimento contínuo a cada atualização de tela, gerando uma curva parabólica natural durante a rasterização.

### 3.2. Espaço do Mundo e Volumes Envolventes (AABB)

Para garantir que os polígonos não se cruzem indevidamente na tela (efeito de *clipping*), o sistema opera dentro do "Espaço do Mundo" (World Space) delimitado pela nossa projeção ortogonal (`gluOrtho2D`).

A técnica de colisão AABB (*Axis-Aligned Bounding Box*) atua como um teste geométrico de volume envolvente. Antes de despachar os vértices para a placa de vídeo, o algoritmo compara as coordenadas limites (largura e altura) do retângulo do jogador com os retângulos do cenário. Caso o cálculo aponte uma intersecção no espaço coordenado, a transformação de translação pretendida é bloqueada, mantendo a integridade visual da cena.

### 3.3. A Câmera Virtual (Translação Inversa de Matriz)

Em um ambiente OpenGL bidimensional puro, não existe uma "câmera" física instanciada. Para cumprir o requisito de uma câmera dinâmica que acompanha o jogador (formato 2.5D), manipulamos diretamente a matriz de visualização e modelagem (`GL_MODELVIEW`).

Isso foi obtido aplicando a função `glTranslatef()` com o valor inverso da coordenada X do protagonista pouco antes de desenhar o cenário. Matematicamente, o Conde permanece estático na coordenada de projeção da tela, enquanto todos os polígonos do ambiente sofrem uma transformação de translação na direção oposta, criando a ilusão perfeita de uma câmera que "segue" o personagem.

### 3.4. Atmosfera Visual: Névoa Fria Noturna (Recurso Gráfico Extra - Em Desenvolvimento)
Para enriquecer a identidade visual e antecipar os requisitos estéticos do trabalho, foi implementado um sistema dinâmico de névoa no protótipo. A técnica utilizada evita "bordas duras" geométricas e gera uma ilusão de profundidade através de três conceitos gráficos fundamentais:

Mistura de Cores (Alpha Blending): O motor ativa a função de transparência do OpenGL (GL_BLEND) utilizando a diretiva glBlendFunc(GL_SRC_ALPHA, GL_ONE_MINUS_SRC_ALPHA).

Gradiente Linear de Opacidade: Para simular a dispersão natural da fumaça, cada faixa de névoa é renderizada combinando polígonos com transição de Alpha nulo (0.0f) para máximo (0.15f)

Parallax Bidimensional (2D): A névoa é composta por camadas independentes que transladam em velocidades distintas, enganando a percepção do espectador e gerando profundidade.


### 4. **Arquitetura da Lógica de Jogo e Regras**

Nesta seção, formalizamos a lógica computacional que rege a interação do jogador com o mundo, elevando o conceito de "fuga da luz" para um sistema dinâmico de intersecção de áreas.

### 4.1. Máquina de Estados e Sistema de Energia

O ciclo de vida do jogo é gerenciado por uma máquina de estados finitos (`MENU`, `JOGANDO`, `GAME_OVER`). Durante o estado `JOGANDO`, o protagonista (Conde) não possui "vidas" binárias, mas sim uma variável contínua chamada `energia_vital` (iniciando em 100%).

A pontuação final do jogador é calculada por uma função que pondera a distância máxima alcançada no eixo X e a quantidade de `energia_vital` restante ao cruzar a linha de chegada da fase.

### 4.2. Algoritmo da Mecânica de Luz e Sombra (Núcleo do Jogo)

A mecânica obrigatória de luz/sombra não se baseia em morte instantânea, mas sim em um cálculo de intersecção de três camadas (Jogador, Luz e Obstáculo).

* **Camada de Risco (Feixes de Luz):** O sistema renderiza polígonos amarelos com transparência (Alpha Blending). Cada feixe possui sua própria *Bounding Box* (`AABB_Luz`).
* **Camada de Proteção (Sombra Dinâmica):** O cenário possui blocos de pedra empurráveis (`AABB_Pedra`). Quando uma pedra está posicionada sob um feixe de luz, o motor de jogo calcula e projeta uma "Zona Segura" (`AABB_Sombra`) diretamente abaixo ou na diagonal do bloco.
* **A Lógica de Dano:** A cada *frame* da física, o sistema realiza a seguinte verificação:
* **SE** (`AABB_Conde` intersecta `AABB_Luz`) **E NÃO** (`AABB_Conde` intersecta `AABB_Sombra`) $\rightarrow$ Aplica-se uma penalidade contínua: $E = E - (dano \times \Delta t)$.
* Caso a `energia_vital` chegue a zero, o estado transita para `GAME_OVER`, acionando o sistema de partículas (transformação em cinzas).



### 4.3. Nível de Dificuldade e Design de Fases (Level Design)

A lógica de progressão baseia-se no aumento da complexidade das variáveis de ambiente:

1. **Fase 1 (A Cidade):** Ambiente de tutorial. Os feixes de luz são estáticos e verticais. O desafio recai apenas sobre a sincronia do pulo para não cair nos buracos abertos nas ruas e evitar as luzes paradas.
2. **Fase 2 (Floresta e Ruínas):** Introdução da mecânica de quebra-cabeça físico. O jogador precisa utilizar a força de translação no eixo X para empurrar pedras até pontos específicos do cenário, criando Zonas Seguras de sombra temporária para conseguir atravessar trechos longos de sol. Introdução de inimigos (animais silvestres) que causam dano por colisão simples (redução instantânea de energia e aplicação de recuo/knockback).
3. **Fase 3 (As Catacumbas):** O ápice lógico. Os feixes de luz deixam de ser estáticos e passam a utilizar matrizes de translação para oscilar no eixo X (luzes que "varrem" o cenário). O jogador precisa combinar o pulo preciso com a leitura do tempo (timing) das luzes móveis, lidando simultaneamente com a perseguição de um Caçador de Vampiros.

### 4.4. Especificação dos Recursos Visuais e Câmeras

Para garantir a percepção da perspectiva 2.5D e imersão:

* **Sistema Multicâmera:**
* `Cam_Follow`: Câmera principal. Utiliza translação inversa no eixo X baseada nas coordenadas do jogador, mantendo-o no terço esquerdo da tela para garantir tempo de reação.
* `Cam_Tática`: Acionada por tecla específica, afasta o zoom manipulando os limites geométricos da projeção ortogonal (`gluOrtho2D`), permitindo o planejamento da rota.


* **Recursos Gráficos Adicionais:**
1. **Parallax Simples:** O fundo (background) transladará a uma velocidade inferior à do plano de jogo, gerando ilusão de profundidade (eixo Z falso).
2. **Partículas Cinemáticas:** Emissão de pequenos quadrados escuros com vetores de dispersão aleatória para simular cinzas ao perder o jogo.
3. **Iluminação de Feedback:** As áreas de sombra seguras terão um leve escurecimento no canal RGB (0.3f, 0.3f, 0.3f) para indicar visualmente ao jogador onde é seguro pisar.
