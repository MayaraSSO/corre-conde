## **1. Introdução e Contextualização**
Este projeto consiste no desenvolvimento de um jogo digital do gênero plataforma 2.5D, desenvolvido como requisito avaliativo para a disciplina de Computação Gráfica. O objetivo central é a aplicação prática de conceitos fundamentais da área, como transformações geométricas, modelos de iluminação, projeções e tratamento de colisões, integrando-os em uma experiência interativa e funcional.

#### **1.1. O Projeto “Corre Conde: A Sombra da Eternidade”**

A narrativa acompanha o **Conde Drácula** após uma batalha clímax contra seu nêmesis, **Van Helsing**. No embate, o caçador destruiu o **"Anel de Sombras"**, a relíquia milenar que protegia o vampiro da luz solar. O anel foi esfacelado em três fragmentos, entregues a três guardiões distintos. Diferente de jogos de plataforma tradicionais, o foco do Corre Conde é a sobrevivência. O personagem deve navegar por cenários variados antes do amanhecer, utilizando a manipulação de sombras para evitar a exposição letal à luz, enquanto utiliza mecânicas de **defesa (Capa Mágica)** e **contra-ataque (Olhar Hipnotizante e Drenagem)** para recuperar os fragmentos e retornar ao seu castelo.

#### **1.2. Objetivos da Etapa 1** 

Nesta fase inicial de desenvolvimento, o foco residiu na estabilização do Protótipo Técnico. As metas estabelecidas e cumpridas nesta etapa (já implementadas no código) incluem:

* > Implementação da base de movimentação horizontal e sistema de salto via integração da física.
  
* > Desenvolvimento de um motor físico básico para detecção de colisões (AABB) com o cenário.

* > Configuração de uma câmera dinâmica que acompanha o deslocamento do protagonista via translação inversa de matriz.

* > Criação de um bloco jogável preliminar com obstáculos e plataformas estáticas para validação das mecânicas.

* > Implementação antecipada de recurso gráfico extra: Sistema de névoa fria noturna com efeito Parallax e interpolação de opacidade (Alpha Blending).


## **2. Especificações Técnicas**
Nesta seção, detalhamos o ecossistema de desenvolvimento utilizado para garantir o desempenho e a portabilidade do protótipo.


#### **2.1 Ferramentas e Bibliotecas**

Para o controle total sobre o pipeline gráfico e gerenciamento eficiente de memória (essencial para sistemas de tempo real), foram escolhidas as seguintes tecnologias:

* Linguagem de Programação: C (C99). Escolhida por permitir o acesso direto ao hardware e por ser o padrão acadêmico para o ensino de Computação Gráfica na Ciência da Computação.
  
* Interface de Programação de Aplicações (API): OpenGL. Utilizada para a renderização de gráficos 2D e 3D, aproveitando a aceleração de hardware.

  
* Toolkit de Utilidades: FreeGLUT. Versão de código aberto do OpenGL Utility Toolkit
(GLUT), responsável pela criação da janela no sistema X11 do Linux, manipulação de buffers e captura de eventos de teclado e mouse de forma simplificada.


* Compilador: GCC (GNU Compiler Collection). Utilizado para a compilação do código fonte através do terminal Linux com o uso das flags -lGL -lGLU -lglut.

  

#### **2.2 Ambiente de Desenvolvimento (Host)**

O desenvolvimento e a depuração do protótipo foram realizados sob as seguintes especificações:

- Sistema Operacional: Ubuntu 24.04.4 LTS (Noble Numbat).
  
- Hardware: Dell Vostro 3460 (Intel Core i3-3110M, 8GB RAM). A configuração validou a execução do motor travada a ~60 FPS em rotinas `glutTimerFunc`, sem perda de ciclos.
  



## **3. Transformações Geométricas e Interação Espacial**

Nesta etapa, o motor do jogo foi estruturado com foco na manipulação do sistema de coordenadas virtuais e na aplicação de transformações geométricas contínuas, que são os pilares da Computação Gráfica bidimensional.

#### **3.1. Animação e Matrizes de Translação (Movimento e Salto)**

No contexto do pipeline gráfico, a movimentação do Conde e a aplicação da gravidade não são  tratadas apenas como grandezas físicas, mas sim como Transformações de Translação.
A cada ciclo de renderização (frame), o sistema calcula os novos deslocamentos nos eixos X e Y. Em vez de utilizar funções complexas de matrizes para cada frame, o motor atualiza diretamente as coordenadas espaciais que compõem os vértices (`glVertex2f`) dos polígonos do personagem (`GL_QUADS`). O pulo é renderizado como uma translação positiva no eixo Y que sofre um decaimento contínuo a cada atualização de tela, gerando uma curva parabólica natural durante a rasterização.


#### **3.2. Espaço do Mundo e Volumes Envolventes (AABB)**

Para garantir que os polígonos não se cruzem indevidamente na tela (efeito de clipping), o sistema opera dentro do **“Espaço do Mundo” (World Space)** delimitado pela nossa projeção ortogonal (`gluOrtho2D`). A técnica de colisão **AABB** (Axis-Aligned Bounding Box) atua como um teste geométrico de volume envolvente. Antes de despachar os vértices para a placa de vídeo, o algoritmo compara as coordenadas limites (largura e altura) do retângulo do jogador com os retângulos do cenário. Caso o cálculo aponte uma intersecção no espaço coordenado, a transformação de translação pretendida é bloqueada, mantendo a integridade visual da cena.

#### **3.3. A Câmera Virtual (Translação Inversa de Matriz)**

Em um ambiente OpenGL bidimensional puro, não existe uma “câmera” física instanciada. Para cumprir o requisito de uma câmera dinâmica que acompanha o jogador (formato 2.5D), manipulamos diretamente a matriz de visualização e modelagem (`GL_MODELVIEW`).
Isso foi obtido aplicando a função `glTranslatef()` com o valor inverso da coordenada X do protagonista pouco antes de desenhar o cenário. Matematicamente, o Conde permanece estático na coordenada de projeção da tela, enquanto todos os polígonos do ambiente sofrem uma transformação de translação na direção oposta, criando a ilusão perfeita de uma câmera que “segue” o personagem.

#### **3.4. Atmosfera Visual**

**Névoa Fria Noturna (Recurso Gráfico Extra - Em Desenvolvimento):** Para enriquecer a identidade visual e antecipar os requisitos estéticos do trabalho, foi implementado um sistema dinâmico de névoa no protótipo. A técnica utilizada evita “bordas duras” geométricas e gera uma ilusão de profundidade através de três conceitos gráficos fundamentais:

**Mistura de Cores (Alpha Blending):** O motor ativa a função de transparência do OpenGL (`GL_BLEND`) utilizando a diretiva glBlendFunc(`GL_SRC_ALPHA`, `GL_ONE_MINUS_SRC_ALPHA`).

**Gradiente Linear de Opacidade:** Para simular a dispersão natural da fumaça, cada faixa de névoa é renderizada combinando polígonos com transição de Alpha nulo (0.0f) para máximo (0.15f).

**Parallax Bidimensional (2D):** A névoa é composta por camadas independentes que transladam em velocidades distintas, enganando a percepção do espectador e gerando profundidade.



## **4. Arquitetura da Lógica de Jogo e Regras**

Nesta seção, formalizamos a lógica computacional que integra o roteiro às mecânicas geométricas de intersecção.

#### **4.1. Máquina de Estados e Arsenal do Protagonista**

O ciclo é regido por uma máquina de estados finitos. O Conde possui a variável `energia_vital` (100%) e acesso a três mecânicas ativas:

* **Capa Mágica:** Modificador de hitbox que anula dano de projéteis.

* **Olhar Hipnotizante:** Gatilho ativado durante as janelas de vulnerabilidade (recarga) dos inimigos.

* **Drenagem de Sangue:** Ação de execução em inimigos hipnotizados, resultando na recuperação de 100% da energia_vital.
  
#### **4.2. Algoritmo da Mecânica de Luz, Sombra e Dano**

O ambiente dita a sobrevivência através de testes contínuos:

* **Camada de Risco Ambiental:** Contato com Alhos, Crucifixos e Quebra de caixão geram uma penalidade para cada um -15% da `energia_vital`. 

* **Camada de Risco de Luz:** Feixes com transparência (`AABB_Luz`). A exposição contínua gera decaimento: $E_{nova} = E_{atual} - (\text{dano} \times \Delta t)$.

* **Camada de Proteção:** Zonas calculadas geometricamente através da translação de rochas ou abrigos estruturais (`AABB_Sombra`) anulam o dano da luz.

#### **4.3. Design de Fases (Level Design) e Progressão**
A lógica de progressão foi estruturada para garantir que o sucesso em cada fase dependa da adaptação do jogador aos perigos do ambiente e do gerenciamento preciso da variável energia_vital:

* > **Fase 1 (A Cidade - Tutorial):** O cenário apresenta feixes de luz estáticos e verticais em oposição à área escura e segura (Névoa). O desafio baseia-se na sincronia de saltos para evitar buracos e alhos no cenário. Ao final, ocorre o combate com o **Paladino**, que utiliza uma balestra com 4 estacas. A intersecção geométrica do protagonista com cada estaca resulta em uma penalidade de **10%** na `energia_vital`. A vitória sobre o chefe garante a coleta do 1º fragmento do anel.

* > **Fase 2 (A Floresta - Puzzles Físicos):** Introdução das Pedras Mágicas. A translação dessas rochas no eixo X gera acúmulo vertical por colisão, projetando novas Bounding Boxes de sombra para proteção. O solo contém crucifixos espalhados; caso o jogador intercepte suas coordenadas, sofrerá uma penalidade instantânea de **15%** da `energia_vital`. O combate de transição é contra o **Lobisomem**, cujo padrão é o ataque físico corpo a corpo contínuo. Cada toque (colisão AABB) do inimigo subtrai **10%** da `energia_vital` do Conde. O jogador deve resistir aos ataques e esquivar por 10 segundos ininterruptos para abrir a janela de hipnose, drenar todo o sangue do **Lobisomem** e, assim, coletar o 2º fragmento.
  
* > **Fase 3 (As Catacumbas/Cemitério - Clímax):** O ambiente insere **caixões** quebradiços distribuídos de forma aleatória; pisar neles causa uma quebra de terreno e uma punição de **15%** na `energia_vital`. A dificuldade atinge o ápice gráfico: as matrizes de translação da luz passam a oscilar dinamicamente no eixo X (luzes de varredura). Simultaneamente, o "amanhecer" empurra a Névoa Segura para a frente de forma contínua, atuando como um cronômetro opressor. O chefe final é **Van Helsing**, que atira 4 projéteis de prata rápidos (cada acerto retira **15%** da `energia_vital). A sobrevivência exige o uso estratégico da Capa de Defesa e transições precisas de eixo (pulos de esquiva). A vitória contra o caçador recupera o último fragmento, permitindo que o Conde alcance seu **objetivo final: a zona de proteção absoluta dentro de seu castelo, concluindo o jogo**.

#### **4.4. Especificação dos Recursos Visuais e Câmeras**

Para garantir a percepção de profundidade na perspectiva 2.5D, a imersão e o controle tático do jogador sobre o espaço virtual, o motor gráfico implementa um **Sistema Multicâmera baseado na manipulação dinâmica das matrizes de projeção e visualização:**

* **Cam_Follow (Câmera Dinâmica Principal):** Atua como o estado padrão de navegação. Utiliza translação inversa no eixo X aplicada diretamente à matriz `GL_MODELVIEW`, calculada em tempo real com base nas coordenadas de deslocamento do jogador. Matematicamente, a matriz é configurada para manter o protagonista sempre fixo no terço esquerdo da área de projeção (viewport), garantindo tempo de reação hábil contra feixes de luz e obstáculos que surgem à direita da tela.
  
* **Cam_Tática (Câmera de Planejamento Espacial):** Acionada através de um evento específico de teclado (capturado via API do FreeGLUT). Esta câmera altera o estado da matriz de projeção geométrica, manipulando os limites de rederização da função `gluOrtho2D`. Ao expandir as coordenadas topológicas da janela ortogonal, o sistema gera um afastamento (zoom out) em tempo real, ampliando o campo de visão sobre o World Space e permitindo que o jogador analise padrões de luzes dinâmicas e planeje rotas seguras.

* **Cam_Batalha (Câmera de Confinamento/Arena):** Acionada de forma automática por um gatilho de intersecção espacial (Trigger geométrico) assim que a Bounding Box do jogador adentra as coordenadas limítrofes do chefe de fase. Este estado interrompe o cálculo contínuo de `Cam_Follow` e fixa a translação da matriz de visualização no centro geográfico da arena de combate. Essa trava garante que tanto o Conde quanto o oponente (Paladino, Lobisomem ou Van Helsing) permaneçam visíveis e contidos dentro das fronteiras geométricas da tela, eliminando pontos cegos durante a resolução dos confrontos e mecânicas de esquiva.

**Recursos Gráficos Adicionais:**

1. **Parallax Simples:** O fundo (background) transladará a uma velocidade inferior à do plano de jogo, gerando ilusão de profundidade (eixo Z falso).
   
3. **Partículas Cinemáticas:** Emissão de pequenos quadrados escuros com vetores de dispersão aleatória para simular cinzas ao perder o jogo.

   
5. **Iluminação de Feedback:** As áreas de sombra seguras terão um leve escurecimento no canal RGB (0.3f, 0.3f, 0.3f) para indicar visualmente ao jogador onde é seguro pisar.
